**function to take dataset and model**

```python
def different_mods(model,dataset):
    labels=model.fit_predict(dataset)
    sns.scatterplot(data=dataset,x='X1',y='X2',hue=labels,palette='Set1')
```

# KMeans

```python
from sklearn.cluster import KMeans
```

```python
model_kmeans=KMeans(n_clusters=2,n_init=10)
```

```python
different_mods(model_kmeans,circles)
plt.title('Kmeans on circles dataset');
```

```python
different_mods(model_kmeans,moons)
plt.title('Kmeans on moons dataset');
```

```python
model_kmeans=KMeans(n_clusters=3,init='k-means++',n_init=10)
different_mods(model_kmeans,bloobs)
plt.title('Kmeans on Bloobs dataset');
```

# Now Let's predict clusters thorugh DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
E=0.15
min_pts=6
```

```python
Dbscan_model=DBSCAN(eps=E, min_samples=min_pts)
```

```python
different_mods(Dbscan_model,moons)
plt.title('DBSCAN on moons dataset');
```

```python
E=1
min_pts=6
Dbscan_model=DBSCAN(eps=E, min_samples=min_pts)
different_mods(Dbscan_model,bloobs)
plt.title('DBSCAN on Bloobs dataset');
```

```python
E=0.3
min_pts=6
Dbscan_model=DBSCAN(eps=E, min_samples=min_pts)
different_mods(Dbscan_model,circles)
plt.title('DBSCAN on circles dataset');
```
