# Visualization

```python
plt.figure(figsize=(8,4))
sns.histplot(df,x='age',bins=30,hue='loan')
```

```python
x=pd.get_dummies(df,drop_first=True)
```

```python
corr=x.corr()
```

```python
sns.pairplot(df)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=2,n_init=10)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
clusters_labels=kmeans.fit_predict(x_scaled)
```

```python
clusters_labels
```

```python
x['clusters']=clusters_labels
```

# correlation of clusters labels

```python
plt.figure(figsize=(10,5))
x.corr()['clusters'].iloc[:-1].sort_values(ascending=True).plot(kind='bar')
```

# Choosing Right value of K

```python
wcss=[]
for k in range(1,50):
    kmeans=KMeans(n_clusters=k,n_init=10)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
    
```

```python
plt.plot(range(1,50),wcss)
```

```python
from kneed import KneeLocator
```

```python
kl=KneeLocator(range(1,20),wcss,curve='convex')
```
