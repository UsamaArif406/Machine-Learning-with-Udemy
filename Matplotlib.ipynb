```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=2*x
a=2*x+4;
b=x**4
```

```python
plt.plot(x,y)
plt.title('Line Grpah')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

# Figures and Axes

```python
fig=plt.figure(figsize=(4,3),dpi=100)
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y);

axes.set_title('Linear Graph using figures and axes')

axes1=fig.add_axes([0.19,0.49,.4,.4])
axes1.plot(a,b);
fig.savefig('line graph',bbox_inches='tight')
```

# Subplots

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,y)
axes[1][0].plot(a,b)
axes[1][1].plot(x,b)
plt.show()
```

# Legends and colors

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,a*x+b,color='#121207',label='x vs a*x+b', marker='*',markersize='12')
ax.plot(a,a**b,color='#721278',label='a vs a^b',marker='o')
ax.legend()
```
