# apply method

```python
def last_four(number):
    return str(number)[-4:]
```

```python
df['last four digits']=df['CC Number'].apply(last_four)
```

```python
df
```

```python
def dollar_sign(bill):
    if bill>=20:
        return '$$$'
    elif bill >=10:
        return '$$'
    else:
        return '$'
```

```python
df['dollars']=df['total_bill'].apply(dollar_sign)
```

```python
df.head()
```

```python
def generosity(total_bill,tip):
    if tip/total_bill>=0.2:
        return 'Generous'
    else:
        return 'Non Generous'
```

```python
df['Generosity']=np.vectorize(generosity)(df['total_bill'],df['tip'])
```
