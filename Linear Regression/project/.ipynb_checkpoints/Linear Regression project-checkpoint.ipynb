```python
df=pd.read_csv('without_outliers_housing_data.csv')
```

# Data cleaning

```python
df.isna().sum()
```

```python
def nan_percentage():
    nan_per=100*df.isna().sum()/len(df)
    nan_per=nan_per[nan_per>0]
    return nan_per
```

```python
nan_per=nan_percentage().sort_values(ascending=False)
nan_per
```

# visualizing most nan values

```python
plt.figure(figsize=(12,6))
sns.barplot(x=nan_per.index,y=nan_per)
plt.xticks(rotation=90);
```

# Dropping rows where nan values are < 1%

```python
df=df.dropna(axis=0,subset=['Mas Vnr Area','Mas Vnr Type','Bsmt Full Bath','Total Bsmt SF','Bsmt Unf SF','Garage Cars','Garage Area','BsmtFin SF 2','BsmtFin SF 1','Electrical'])
```

```python
nan_per=nan_percentage().sort_values(ascending=False)
nan_per
```

# Filling nan values where % is >1% and <80%

```python
str_col=['Garage Finish','Garage Qual','Garage Type','Garage Cond']
df[str_col]=df[str_col].fillna('none')
```

```python
df['Garage Yr Blt']=df['Garage Yr Blt'].fillna(0)
```

```python
bsmt_cols=['Bsmt Exposure','BsmtFin Type 2','Bsmt Qual','Bsmt Cond','BsmtFin Type 1','Fireplace Qu']
df[bsmt_cols]=df[bsmt_cols].fillna('none')
```

```python
nan_per=nan_percentage().sort_values(ascending=False)
nan_per
```

```python
df['Lot Frontage']=df['Lot Frontage'].fillna(df['Lot Frontage'].mean())
```

```python
nan_per=nan_percentage().sort_values(ascending=False)
nan_per
```

```python
df=df.drop(['Pool QC','Misc Feature','Alley','Fence','PID'],axis=1)
```

```python
df.head()
```

```python
df=df.drop(['Unnamed: 0','Order'],axis=1)
```

```python
df.head()
```

# Dealing with categorical Features

```python
df['MS SubClass']=df['MS SubClass'].astype('str')
```

```python
object_df=df.select_dtypes(include='object')
numeric_df=df.select_dtypes(exclude='object')
```

```python
dummy_variables=pd.get_dummies(object_df,drop_first=True)
```

```python
final_df=pd.concat([dummy_variables,numeric_df],axis=1)
```

```python
final_df.columns
```

# Splitting data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x=final_df.drop('SalePrice',axis=1)
y=final_df.SalePrice
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
scaler=StandardScaler()
scaler.fit(x_train);
```

```python
x_train=scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elastic_model=ElasticNet(max_iter=1000000)
```

```python
param_grid={'alpha':[0.1,1,2,3,4,100,150],'l1_ratio':[0.1,0.7,.6,.99,1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model=GridSearchCV(estimator=elastic_model,param_grid=param_grid,cv=5,
                        scoring='neg_mean_squared_error')
```

```python
grid_model.fit(x_train,y_train)
```

```python
grid_model.best_params_
```

```python
y_pred=grid_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE=mean_absolute_error(y_test,y_pred)
```

```python
MAE
```

```python
RMSE=np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
RMSE
```
