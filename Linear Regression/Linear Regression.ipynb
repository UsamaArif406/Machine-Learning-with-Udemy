```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('advertising.csv')
```

```python
df.head()
```

```python
df['total_spend']=df['TV']+df['Radio']+df['Newspaper']
```

```python
df.head()
```

# Visuallization

```python
sns.regplot(data=df,x='total_spend',y='Sales')
```

# We can plot this best fitt line by first calculating Beta values and putting them in Y=B1x+B0

first separate features and labels 

```python
X=df.total_spend
Y=df.Sales
```

```python
B1,B0=np.polyfit(X,Y,deg=1)
```

# let's plot best fit line through the datapoints

```python
potential_spend=np.linspace(0,450,100)
```

```python
#we know y=B1*x+B0 
y_preds=B1*potential_spend+B0
```

```python
sns.scatterplot(data=df,x='total_spend',y='Sales',color='green')
plt.plot(potential_spend,y_preds,color='red')
plt.show()
```
