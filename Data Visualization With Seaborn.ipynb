# Distribution Plots

# rug plots

A rug plot is a simple way to visualize data points along a single axis. It places a small vertical tick at each data point's position

```python
sns.rugplot(data=df,x='salary',height=0.2,hue='division')
```

# Histplot (Histogram )Plot

A histplot is used to create a histogram of your data. It divides the data into bins and displays the frequency of data points within each bin.

```python
sns.histplot(data=df,x='salary',bins=30,color='crimson',hue='level of education');
```

A KDE plot is used to estimate the probability density function of a continuous random 

```python
sns.kdeplot(data=df,x='salary',hue='division',thresh=0.5);
```

```python
sns.displot(data=df,x='salary',kde=True,rug=True,hue='division')
```

```python
ds=pd.read_csv('StudentsPerformance1.csv')
ds.head()
```

# BoxPlot

# Box
The box represents the interquartile range (IQR), which includes the middle 50% of the data. The top and bottom of the box represent the 75th percentile (Q3) and the 25th percentile (Q1), respectively.

# Line (Whiskers)
The lines extending from the box, known as whiskers, show the range of the data within a specified distance from the quartiles. Often, this distance is set to 1.5 times the IQR. Data points outside this range are considered potential outliers and are typically plotted individually.

# Outliers
Individual data points that fall outside the whiskers are often plotted as individual points to draw attention to them

# lets explore this data with box plot

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=ds,x='parental level of education',y='writing score',hue='gender');
plt.legend(bbox_to_anchor=(1.5,0.5))
plt.xticks(rotation=10);
```

# Violin plot
# Box
Just like in a box plot, the box in a violin plot represents the interquartile range (IQR).

# Violin Shape
The main part of the plot is the "violin" shape. This shape represents the probability density of the data at different values, similar to a probability distribution.

# Two Lines
The two lines within the violin plot represent the median (center line) and quartiles (if visualized) of the data. These lines provide a summary of the central tendency and spread of the data

```python
plt.figure(figsize=(8,5))
sns.violinplot(data=ds,x='lunch',y='math score',hue='gender');
plt.legend(bbox_to_anchor=(1.5,0.5))
plt.xticks(rotation=10);
```

# swarmplot

# Categories 
A swarm plot typically visualizes data points from one or more categories or groups along a single axis. Each category is represented by a vertical column of points.

# Individual Data Points
Each data point within a category is plotted as a single point. These points are spread out horizontally within the category's column. Horizontal jittering may be applied to prevent overlapping points.

# No Overlapping
One of the primary features of swarm plots is that they prevent data points from overlapping within a category. This makes it easier to see the density and distribution of the data within each category.

# Use of Color
Color can be used to distinguish different categories if there are multiple categories on the same axis.

```python
plt.figure(figsize=(8,5))
sns.swarmplot(data=ds,x='math score',y='gender');
```

# Comparison Plots

# Jointplot 
is used to create a multi-panel plot that visualizes the relationship between two variables. It shows both the individual variable distributions along the axes and the joint relationship between the two variables in the center. This is helpful for understanding the correlation, distribution, and outliers in your data.

```python
sns.jointplot(data=ds,x='math score',y='writing score',kind='kde',palette='deep',hue='gender',fill=True);
```

# Pairplot 
is used to create a grid of scatterplots that shows the pairwise relationships between multiple variables in a dataset. This can be very useful for identifying trends, correlations, and patterns between different pairs of variables in your data.

```python
sns.pairplot(data=ds,diag_kind='kde',hue='gender',palette='deep')
```

```python
sns.catplot(data=ds,x='math score',y='writing score',col='gender',row='test preparation course',palette='dark')
```

# Matrix plot

```python
data=pd.read_csv('Country-data.csv')
data=data[110:125]
data.head(10)
```

```python
data=data.set_index('country')
```

# Heatmap
is a data visualization technique that represents data in a two-dimensional grid format using colors. It is a way to display the magnitude of a numerical value as color in a matrix or grid of cells. Heatmaps
are commonly used for various purposes, such as visualizing data correlations, density, or patterns within large datasets.

```python
plt.figure(figsize=(12,5))
sns.heatmap(data,annot=True,linewidths=0.5,cmap='coolwarm');
```

# Cluster map
is a type of heatmap that is used to visualize the results of a hierarchical clustering analysis. Hierarchical clustering is a method of grouping similar data points or elements into clusters or dendrograms based on their similarity or dissimilarity

```python
plt.figure(figsize=(12,5))
sns.clustermap(data,annot=True,linewidths=0.5,cmap='viridis');
plt.show()
```
