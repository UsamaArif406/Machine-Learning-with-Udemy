```python
country_sales1 = {
    "United States": 1000,
    "Canada": 5000,
    "United Kingdom": 7500,
    "Australia": 6600,
    "Germany": 8500
}
country_sales2={
    "United States": 20000,
    "Pakistan": 25000,
    "SriLanka": 7000,
    "Australia": 62000,
    "Germany": 85200
}
```

```python
sales_1=pd.Series(country_sales1,name='country_sales')
sales_2=pd.Series(country_sales2,name='country_sales')
```

```python
sales_1.add(sales_2,fill_value=0)
```

# in above Example
we have two Series, sales_1 and sales_2, with some common and some unique index values. When we use the add method with fill_value=0, it adds the corresponding values where the indices match and fills in 0 for indices that are missing in either of the Series. The result Series will contain the sum of the values for each index.

# Dropping NAN values

```python
record=sales_1.add(sales_2)
record
```

```python
#record.dropna(inplace=True)
```

Or simply we can drop NAN values as done in above example or can replace them with 0 or anay number we want (see example below)

# groupby
The groupby method in pandas is used to split data into groups based on specified criteria, apply functions to these groups, and then combine the results. It's commonly used for data aggregation and analysis.It is applied over categorical column

```python
cars=pd.read_csv('auto-mpg.csv')
cars.head()
```

```python
model_year=cars.groupby('model year')
```

```python
#we have splitted the data now lets apply a mean fucntion over it
model_year.mean()
```

```python
max_val=model_year.max()
```

```python
max_val.loc[81] #max value for 1981 model for different specs
```

```python
#we can apply groupby on data with the help of conditional filtering
cars[cars['cylinders'].isin([4,8])].groupby(['model year','cylinders']).mean()
```

# agg Method

```python
#with the help of agg method we apply several statistics approaches on svevral columns
cars.agg({'mpg':['mean','max'],'weight':['mean','std']})
```

# Merge
# Inner
An inner join combines two datasets, keeping only the matching rows from both sets, leaving out non-matching rows

```python
log_data = {
    'Name': ['Xvaier', 'Azlan', 'Ali', 'Abubakar'],
    'Login ID': [101, 102, 103, 104]
}

log_df = pd.DataFrame(login_data)

reg_data = {
    'Name': ['Andrew', 'Azlan', 'Ali', 'sam'],
    'Registration ID': [501, 502, 503, 504]
}

reg_df = pd.DataFrame(registration_data)
```

```python
df=pd.merge(log_df,reg_df,how='inner',on='Name')
```

```python
df
```

# Strings Method

```python
mails=['S; am 1@em! ail.com',
'john l;m ail.c!om',
'fredric; k@ema il.co!m',
'joseph@   ;mail.com',
 'christ @email.com']
```

```python
series=pd.Series(mails)
```

```python
series
```

```python
#lets clean them up using apply method first
def cleanup(mails):
    mails=mails.capitalize()
    mails=mails.replace(' ','')
    mails=mails.replace(';','')
    mails=mails.replace('!','')
    return mails

    
```

```python
series=series.apply(cleanup)
```

```python
series
```

```python
#this can be done using pandas str methos aswell and take less time
series.str.replace(' ','').str.replace(';','').str.replace('!','')
```
